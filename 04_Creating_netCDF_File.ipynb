```python
# Adding data variable

import numpy as np


depths = [0, 10, 20, 25]
latitude = [12,45,23]
longitude = [39,42,23]

start = np.datetime64('2024-01-01T00:00:00')
end = start + np.timedelta64(24, 'h')
timestamps = np.arange(start, end, np.timedelta64(1, 'h'))

seconds_since_start = (timestamps - start) / np.timedelta64(1, 's')
seconds_since_start = seconds_since_start.astype(np.float64)

chlorophyll_a = [12.56,45.6,45.5,12] #4 points, one for each depths



xrds = xr.Dataset(
    coords = {
        'depth': depths,
        'latitude': latitude,
        'longitude': longitude,
        'time': seconds_since_start
    },
    data_vars = {
        'chlorophyll_a': ('depth', chlorophyll_a)
    }
)

xrds
```

```python
# Adding data variables having more than one D

import numpy as np


depths = [0, 10, 20, 25]
latitude = [12,45,23]
longitude = [39,42,23]

start = np.datetime64('2024-01-01T00:00:00')
end = start + np.timedelta64(24, 'h')
timestamps = np.arange(start, end, np.timedelta64(1, 'h'))

seconds_since_start = (timestamps - start) / np.timedelta64(1, 's')
seconds_since_start = seconds_since_start.astype(np.float64)

chlorophyll_a = [12.56,45.6,45.5,12] #4 points, one for each depths

wind_speed = np.random.randint(0,10,size=(3,3)) #adding 2d array of wind speed
print(wind_speed)

xrds = xr.Dataset(
    coords = {
        'depth': depths,
        'latitude': latitude,
        'longitude': longitude,
        'time': seconds_since_start
    },
    data_vars = {
        'chlorophyll_a': ('depth', chlorophyll_a),
        'wind_speed': (['latitude', 'longitude'], wind_speed)
    }
)

xrds
```

```python
# Adding 3d Temp data

import numpy as np


depths = [0, 10, 20, 25]
latitude = [12,45]
longitude = [39,42,23]

start = np.datetime64('2024-01-01T00:00:00')
end = start + np.timedelta64(24, 'h')
timestamps = np.arange(start, end, np.timedelta64(1, 'h'))

seconds_since_start = (timestamps - start) / np.timedelta64(1, 's')
seconds_since_start = seconds_since_start.astype(np.float64)

chlorophyll_a = [12.56,45.6,45.5,12] #4 points, one for each depths

wind_speed = np.random.randint(0,10,size=(2,3)) #adding 2d array of wind speed

temp = np.random.randint(0,10,size=(len(latitude), len(longitude), len(depths))) #adding 3d array of temp
print(temp)

xrds = xr.Dataset(
    coords = {
        'depth': depths,
        'latitude': latitude,
        'longitude': longitude,
        'time': seconds_since_start
    },
    data_vars = {
        'chlorophyll_a': ('depth', chlorophyll_a),
        'wind_speed': (['latitude', 'longitude'], wind_speed),
        'temp': (['latitude', 'longitude', 'depth'], temp)
    }
)

xrds
```
